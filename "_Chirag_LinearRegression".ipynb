**Prediction of house prices using Linear regression**

---



The dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015.

```python
#Linear Regression model to find rental price
#I have taken 4 cases of this linear regression model to predict thr price of a rental system
#we basically find the acuracy and try to increase the acuracy of the data set..so that the prediction in spot on
#case1
#regression model is made taking sqft_living into consideration
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn import linear_model
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
dff=pd.read_csv(file_name)
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
dff=pd.read_csv(file_name)
cdf=dff[['sqft_lot','price','bedrooms','sqft_living']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(cdf,cdf,test_size=.2,random_state=4);
regr = linear_model.LinearRegression()
train_x = np.asanyarray(X_train[['sqft_living']])
train_y = np.asanyarray(X_train[['price']])
regr.fit(train_x, train_y)
plt.scatter(dff.sqft_living, dff.price,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'red')
plt.xlabel("sqft_living")
plt.ylabel("price")
plt.show()
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
test_x = np.asanyarray(X_test[['sqft_living']])
test_y = np.asanyarray(X_test[['price']])
test_y_ = regr.predict(test_x)
mae= np.mean(np.absolute(test_y_ - test_y))
mae_=mae/21613
mse=np.mean((test_y_ - test_y) ** 2)
mse_=mse/21613
print("Mean absolute error: " , mae_)
print("Residual sum of squares (MSE): " ,mse_ )
#from the output we can see that the  Mean absolute error:  8.01952622105321
#And Residual sum of squares (MSE):  3093754.956167733
#this is not a great acuracy..
#we need to improve the acuracy of the model using othe kinds of regression(like multivariable regression)

#********************************************************************************************************************

#case 2
#considering view as a criteria to predict the price
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
#dff=pd.read_csv(file_name)
cdf=dff[['date','price','bedrooms','view']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(cdf,cdf,test_size=.2,random_state=4);
regr = linear_model.LinearRegression()
train_x = np.asanyarray(X_train[['view']])
train_y = np.asanyarray(X_train[['price']])
regr.fit(train_x, train_y)
plt.scatter(dff.view, dff.price,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'red')
plt.xlabel("view")
plt.ylabel("price")
plt.show()
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
test_x = np.asanyarray(X_test[['view']])
test_y = np.asanyarray(X_test[['price']])
test_y_ = regr.predict(test_x)
mae= np.mean(np.absolute(test_y_ - test_y))
mae_=mae/21613
mse=np.mean((test_y_ - test_y) ** 2)
mse_=mse/21613
print("Mean absolute error: " , mae_)
print("Residual sum of squares (MSE): " ,mse_ )
#from the output we can see that the  Mean absolute error:  10.000116345300357
# And Residual sum of squares (MSE):  5010836.042746316
#these values are not of great acuracy(the value ideally should be zero(100% accurate) but its too hig in this case)
#we cant use this to predict the price 
#we can drop the columns containing Nan values to increase the accuracy
#dropping can be donw using drop() function

#********************************************************************************************************************

#case 3
#considering floors as a criteria
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
dff=pd.read_csv(file_name)
cdf=dff[['date','price','grade','floors']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(cdf,cdf,test_size=.2,random_state=4);
regr = linear_model.LinearRegression()
train_x = np.asanyarray(X_train[['floors']])
train_y = np.asanyarray(X_train[['price']])
regr.fit(train_x, train_y)
plt.scatter(dff.floors, dff.price,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'red')
plt.xlabel("floors")
plt.ylabel("price")
plt.show()
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
test_x = np.asanyarray(X_test[['floors']])
test_y = np.asanyarray(X_test[['price']])
test_y_ = regr.predict(test_x)
mae= np.mean(np.absolute(test_y_ - test_y))
mae_=mae/21613
mse=np.mean((test_y_ - test_y) ** 2)
mse_=mse/21613
print("Mean absolute error: " , mae_)
print("Residual sum of squares (MSE): " ,mse_ )
#from the output we can see that the Mean absolute error:  10.318775527860838
#And Residual sum of squares (MSE):  5614093.94515966
#these values are of great acuracy(the value ideally should be zero so the one which is very close to 0)
#this case can also be used to predict the price 
#this is not a great acuracy..
#we need to improve the acuracy of the model using othe kinds of regression(like multivariable regression)
#or by dropping some of the columns or transfoemations

#********************************************************************************************************************

#case 4
#considering sqft_lot as a criteria
file_name='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
dff=pd.read_csv(file_name)
cdf=dff[['date','price','grade','sqft_lot']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(cdf,cdf,test_size=.2,random_state=4);
regr = linear_model.LinearRegression()
train_x = np.asanyarray(X_train[['sqft_lot']])
train_y = np.asanyarray(X_train[['price']])
regr.fit(train_x, train_y)
plt.scatter(dff.sqft_lot, dff.price,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'red')
plt.xlabel("sqft_lot")
plt.ylabel("price")
plt.show()
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
test_x = np.asanyarray(X_test[['sqft_lot']])
test_y = np.asanyarray(X_test[['price']])
test_y_ = regr.predict(test_x)
mae= np.mean(np.absolute(test_y_ - test_y))
mae_=mae/21613
mse=np.mean((test_y_ - test_y) ** 2)
mse_=mse/21613
print("Mean absolute error: " , mae_)
print("Residual sum of squares (MSE): " ,mse_ )
#from the output we can see that the Mean absolute error:  10.625294828936248
#And Residual sum of squares (MSE):  5927072.61627958
#these values are of great acuracy(the value ideally should be zero so the one which is very close to 0)
#this case can also be used to predict the price 
#this is not a great acuracy..
#we need to improve the acuracy of the model using othe kinds of regression(like multivariable regression)
#or by dropping some of the columns or transfoemations
#the acuracy also depends on the given data set
#********************************************************************************
#WAYS TO INCREASE DATA SET
#DROPING 
#REPLACING MEAN OR AVERAGE VALUE FOR UNKNOWN TERMS
#USING OTHER KIND OF REGRESSION
#IMPROVING THE DATA SET
#NORMALIZTION OF VALUES
#********************************************************************************
```
